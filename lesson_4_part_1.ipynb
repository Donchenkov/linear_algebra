a)

Найдем ранги матрицы коэфицентов и расширенной матрицы:

Матрица коэффицентов:

```python
a = np.array([[3, -1, 1], [2, -5, -3], [1, 1, -1]])
r = np.linalg.matrix_rank(a)

print(f'Ранг матрицы: {r}')
```

Расширенная матрица:

```python
a = np.array([[3, -1, 1, 4], [2, -5, -3, -17], [1, 1, -1, 0]])
r = np.linalg.matrix_rank(a)

print(f'Ранг матрицы: {r}')
```

$rankA=rank \tilde A=3$, где $3$ — число неизвестных. Следовательно система определена, совместна и имеет единственное решение

б)

Найдем ранги матрицы коэфицентов и расширенной матрицы:

```python
a = np.array([[2, -4, 6], [1, -2, 3], [3, -6, 9]])
r = np.linalg.matrix_rank(a)

print(f'Ранг матрицы: {r}')
```

Расширенная матрица:

```python
a = np.array([[2, -4, 6, 1], [1, -2, 3, -2], [3, -6, 9, 5]])
r = np.linalg.matrix_rank(a)

print(f'Ранг матрицы: {r}')
```

$rankA<rank \tilde A$, система несовместна, решений не имеет.
